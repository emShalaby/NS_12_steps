```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def diffusion(nx):
    dx=2/(nx-1)
    nt=20
    sigma=.2
    dt=sigma*dx
    nu=.3
    dt=sigma*dx**2/nu

    u=numpy.ones(nx)
    u[int(.5/dx):int(1/dx)+1]=2
    un=numpy.ones(nx)

    for n in range(nt):
        un=u.copy()
        for i in range(1,nx-1):
            u[i]=nu*(dt/dx**2)*(un[i+1]-2*un[i]+un[i-1])+un[i]
    plt.plot(numpy.linspace(0,2,nx),u)
diffusion(41)
diffusion(100)
diffusion(1000)
```
