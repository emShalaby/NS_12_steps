```python
import numpy
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
nx=81
ny=81
dx=2/(nx-1)
dy=2/(ny-1)
nt=100
sigma=.2
dt=sigma*dx
c=1

u=numpy.ones((ny,nx))
u[int(.5/dy):int(1/dy+1),int(.5/dx):int(1/dx+1)]=2

v=numpy.ones((ny,nx))
v[int(.5/dy):int(1/dy+1),int(.5/dx):int(.5/dx+1)]=2


```

```python
x=numpy.linspace(0,2,nx)
y=numpy.linspace(0,2,ny)
X,Y=numpy.meshgrid(x,y)

fig=plt.figure(figsize=(11,7),dpi=100)#initilaizing figure size and resolution

ax=plt.axes(projection="3d")


surf=ax.plot_surface(X,Y,u[:],cmap=plt.cm.viridis,rstride=2,cstride=2)#
```

```python
u=numpy.ones((ny,nx))
u[int(.5/dy):int(1/dy+1),int(.5/dx):int(1/dx+1)]=2
un=u.copy()

v=numpy.ones((ny,nx))
v[int(.5/dy):int(1/dy+1),int(.5/dx):int(1/dx+1)]=2
vn=v.copy()
for n in range(nt+1):
    un=u.copy()
    vn=v.copy()
    u[1:,1:]=-vn[1:,1:]*dt/dy*(un[1:,1:]-un[:-1,1:])-un[1:,1:]*dt/dx*(un[1:,1:]-u[1:,:-1])+un[1:,1:]

    v[1:,1:]=-dt/dx*un[1:,1:]*(vn[1:,1:]-vn[1:,:-1])-dt/dy*vn[1:,1:]*(vn[1:,1:]-vn[:-1,1:])+vn[1:,1:]
    
    u[0,:]=1
    u[-1,:]=1
    u[:,0]=1
    u[:,-1]=1
    
    v[0,:]=1
    v[-1,:]=1
    v[:,0]=1
    v[:,-1]=1

fig=plt.figure(figsize=(11,7),dpi=100)
ax=plt.axes(projection="3d")
ax.plot_surface(X,Y,u,cmap=plt.cm.viridis)
```
